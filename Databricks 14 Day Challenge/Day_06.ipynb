```python
oct_events = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
    header=True,
    inferSchema=True
)
nov_events = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv",
    header=True,
    inferSchema=True
)
```

```python
print(f"October 2019 - Total Events: {oct_events.count():,}")
oct_events.printSchema()

print(f"November 2019 - Total Events: {nov_events.count():,}")
nov_events.printSchema()
```

```python
display(oct_events.limit(10))
display(nov_events.limit(10))
```

```python
append_events = oct_events.unionByName(nov_events)
```

```python
append_events.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_events")
```

```python
bronze_events_count=spark.table("bronze_events").count()
spark.table("bronze_events").printSchema()
print(f"Total Events: {bronze_events_count:,}")
display(spark.table("bronze_events").limit(10))
```

```python
from pyspark.sql.functions import col, sum

null_counts = (
    spark.table("bronze_events")
    .select([
        sum(col(c).isNull().cast("int")).alias(c)
        for c in spark.table("bronze_events").columns
    ])
)
null_counts.show(truncate=False)
```

```python
total_rows = spark.table("bronze_events").count()

distinct_rows = spark.table("bronze_events") \
    .select(
        "event_time",
        "user_id",
        "product_id",
        "event_type"
    ) \
    .distinct() \
    .count()

print(f"Total rows     : {total_rows}")
print(f"Distinct rows  : {distinct_rows}")
print(f"Duplicate rows : {total_rows - distinct_rows}")
```

```python
silver_df = spark.table("bronze_events") \
    .dropDuplicates([
        "event_time",
        "user_id",
        "product_id",
        "event_type"
    ])
```

```python
silver_df.groupBy(
    "event_time",
    "user_id",
    "product_id",
    "event_type"
).count().filter("count > 1").show()
```

```python
%sql
DROP TABLE IF EXISTS gold_table;
```

```python
from pyspark.sql.functions import col

gold_clean_df = (
    spark.table("silver_events")
    .filter(col("event_type") == "purchase")
    .filter(col("brand").isNotNull())
    .filter(col("category_code").isNotNull())
)
```

```python
gold_clean_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_layer")
```

```python
spark.table("gold_layer").printSchema()
```

```python
from pyspark.sql.functions import month, year, sum,round

monthwise_revenue = (
    spark.table("gold_layer")
    .withColumn("year", year("event_time"))
    .withColumn("month", month("event_time"))
    .groupBy("year", "month")
    .agg(
        round(sum("price")/1000000,2).alias("total_revenue_million")
    )
    .orderBy("year", "month")
)

display(monthwise_revenue)
```

```python
from pyspark.sql.functions import sum,round,desc

brand_revenue = (
    spark.table("gold_layer")
    .groupBy("brand")
    .agg(
        round(sum("price")/1000000,2).alias("total_revenue")
    )
    .orderBy(desc("total_revenue"))
)
display(brand_revenue.limit(10))
```

```python
event_type_revenue =  spark.table("gold_layer") \
    .groupBy("event_type") \
    .agg(round(sum("price")/1000000,2).alias("total_revenue"))\
    .orderBy(desc("total_revenue"))

display(event_type_revenue)
```

```python
from pyspark.sql.functions import col, sum

gold_df = spark.table("gold_layer")

null_counts = gold_df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in gold_df.columns
])

null_counts.show(truncate=False)
```

```python
from pyspark.sql.functions import count

gold_df = spark.table("gold_layer")

dup_df = (
    gold_df
    .groupBy("event_time", "user_id", "product_id", "user_session")
    .count()
    .filter("count > 1")
)
display(dup_df)
```
