```python
# bronze_layer notebook
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/*.csv")

df.write \
  .mode("append") \
  .saveAsTable("bronze_events")
```

```python
from pyspark.sql.functions import col

df = spark.table("bronze_events")

silver_df = (
    df.dropDuplicates()
      .filter(col("category_code").isNotNull())
      .filter(col("brand").isNotNull())
      .filter(col("user_session").isNotNull())
)
silver_df.write \
  .mode("overwrite") \
  .saveAsTable("silver_events")
```

```python
gold_df = (
    spark.table("silver_events")
    .filter(col("event_type") == "purchase")
)
gold_df.write \
  .mode("overwrite") \
  .saveAsTable("gold_layer")
```
