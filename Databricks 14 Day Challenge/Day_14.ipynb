```python
events = spark.table("ecommerce.silver.silver_events")
```

```python
from pyspark.sql.functions import when

nlp_df = events.withColumn(
    "intent",
    when(events.event_type == "purchase", "high_intent")
    .when(events.event_type == "view", "low_intent")
    .otherwise("medium_intent")
)

nlp_df.groupBy("intent").count().show()
```
